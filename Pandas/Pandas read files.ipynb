```python
from io import StringIO
```

```python
import pandas as pd
```

```python
df = pd.read_csv('mercedesbenz.csv')
```

```python
df.head()
```

```python
df = pd.read_csv('mercedesbenz.csv',usecols=['X0','X1','X3','X4','X5'])
df.head()
```

```python
df.to_csv('test.csv') #saved the data in to new csv file
df.head()
```

```python
df.to_csv('test.csv',index=False)
df.head()
```

```python
# datatypes in csv
data = ('a,b,c,d\n'
        '1,2,3,4\n'
       '5,6,7,8\n'
       '9,10,11')
```

```python
df=pd.read_csv(StringIO(data))
df
```

```python
df.info()
```

```python
df=pd.read_csv(StringIO(data),dtype={'a':int,'b':float,'c':int})
```

```python
df.info()
```

```python
# use index cols and usecols
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data)
```

```python
pd.read_csv(StringIO(data),usecols=['a','b','c'])
```

```python
pd.read_csv(StringIO(data),usecols=['a','b','c'],index_col=0)
```
